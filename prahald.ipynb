```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

```

```python
df=pd.read_csv(r"C:\Users\rishi\ml_projects\Applied-AI-in-Finance\Sapm\data - Sheet1.csv")
df
age=df['Age'].to_numpy()
activity=df['Physical Activity Level'].to_numpy()
heart_rate=df['Heart Rate'].to_numpy()
stress_level=df['Stress Level'].to_numpy()

y_vals=[]
for i in range(len(age)):
    temp=[age[i],activity[i],heart_rate[i]]
    y_vals.append(temp)
y_vals

model=LinearRegression()
model.fit(y_vals,stress_level)
model.coef_[0]

print(f"age :{model.coef_[0]}, activity : {model.coef_[1]}, heart_rate: {model.coef_[2]}")
```
