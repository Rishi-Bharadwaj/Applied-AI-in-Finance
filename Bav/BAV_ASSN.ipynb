```python
import pandas as pd
import matplotlib.pyplot as plt
tax_rate=0.27
ebit_pct_change = ebit.pct_change().mean()
depreciation_pct_change = depreciation.pct_change().mean()
capex_pct_change = capex.pct_change().mean()
wc_delta_pct_change = wc_delta.pct_change().mean()
net_borrowing_pct_change = net_borrowing.pct_change().mean()
interest_expense_pct_change=interest_expense.pct_change().mean()
fcfe_forecast=pd.DataFrame(columns=['Year','EBIT','Depreciation','Capex','Change in Working Capital','Interest Expense','Change in Net Borrowing','FCFF','FCFE'])

for year in ['2022-03-31','2023-03-31','2024-03-31']:
    row1=[year,ebit[year],depreciation[year], capex[year], wc_delta[year],interest_expense[year],net_borrowing[year], fcff[year], fcfe[year]]
    fcfe_forecast.loc[len(fcfe_forecast)] = row1

last_date='2024-03-31'
for i in range(1, 11):
    year = str(int(last_date[:4]) + i) + last_date[4:]
    prev_index = len(fcfe_forecast) - 1  # Index of the last row
    
    forecast_ebit = fcfe_forecast.loc[prev_index, 'EBIT'] * (1 + ebit_pct_change)
    forecast_depreciation = fcfe_forecast.loc[prev_index, 'Depreciation'] * (1 + depreciation_pct_change)
    forecast_capex = fcfe_forecast.loc[prev_index, 'Capex'] * (1 + capex_pct_change)
    forecast_wc_delta = fcfe_forecast.loc[prev_index, 'Change in Working Capital'] * (1 + wc_delta_pct_change)
    forecast_net_borrowing = fcfe_forecast.loc[prev_index, 'Change in Net Borrowing'] * (1 + net_borrowing_pct_change)
    forecast_interest_expense = fcfe_forecast.loc[prev_index, 'Interest Expense'] * (1 + interest_expense_pct_change)
    
    forecast_fcff = forecast_ebit * (1 - tax_rate) + forecast_depreciation - forecast_capex - forecast_wc_delta
    forecast_fcfe = forecast_fcff + forecast_net_borrowing - forecast_interest_expense * (1 - tax_rate)
    fcfe_forecast.loc[len(fcfe_forecast)] = [
        year,
        forecast_ebit,
        forecast_depreciation,
        forecast_capex,
        forecast_wc_delta,
        forecast_interest_expense,
        forecast_net_borrowing,
        forecast_fcff,
        forecast_fcfe
    ]
fcfe_forecast
```

```python
import statsmodels.api as sm
ticker = "CIPLA.NS" 
market_index = "^NSEI" 

start_date = '2019-03-31'
end_date = '2024-03-31'
cipla_data= yf.download(ticker, start=start_date, end= end_date)
market_data= yf.download(market_index, start=start_date,end= end_date)
cipla_returns = cipla_data['Adj Close'].pct_change().dropna()
market_returns = market_data['Adj Close'].pct_change().dropna()

#aligning data by date
returns = pd.concat([cipla_returns, market_returns], axis=1).dropna()
returns.columns = ['Cipla', 'Market']

#REGRESSION ANALYSIS 
X = returns['Market']
y = returns['Cipla']
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
beta_equity = model.params['Market']

print(f"Beta Equity for Cipla: {beta_equity:.2f}")
```

**Computed Beta Equity for CIPLA = 0.40**

```python
plt.figure(figsize=(10, 6))
plt.scatter(X['Market'], y, color='blue', alpha=0.5, label="Data Points")
plt.plot(X['Market'], model.predict(X), color='red', label="Regression Line")
plt.xlabel("Market Returns (NIFTY 50)")
plt.ylabel("Cipla Returns")
plt.title("Beta Calculation for Cipla using Linear Regression")
plt.legend()
plt.show()

print(model.summary())
```

```python
import yfinance as yf 
def get_outstanding_shares(ticker):
    stock_data = yf.Ticker(ticker)
    outstanding_shares = stock_data.info['sharesOutstanding']  # Fetch the number of shares outstanding
    return outstanding_shares

def get_stock_price(ticker):
    stock_data = yf.Ticker(ticker)
    stock_price = stock_data.history(period='1d')['Close'].iloc[-1]  # Get the latest closing price
    return stock_price

ticker = "CIPLA.NS"
outstanding_shares = get_outstanding_shares(ticker)
stock_price = get_stock_price(ticker)
mve = outstanding_shares * stock_price
print(f'Market value of equity is: {mve}')
```

```python
yield_on_debt = 0.0768  # 7.68% yield
time_to_maturity = 5  # 5 years remaining

pv_interest_payments = (income_statement['Interest Expense'] / yield_on_debt) * (1 - (1 / (1 + yield_on_debt) ** time_to_maturity))  
pv_principal = balance_sheet['Total Debt'] / ((1 + yield_on_debt) ** time_to_maturity)

mvd = pv_principal + pv_interest_payments
print(f'Market Value of Debt from Book Value is: {mvd}')
avg_mvd = mvd.mean()
print(f'Average Market Value of Debt is: {avg_mvd}')
```

```python
vfirm = mve + mvd #value of firm
print(f'Value of firm is: {vfirm}')

coc = (mve/vfirm)*cost_of_equity_capital + (mvd/vfirm)*cost_of_debt*(1-tax_rate)
print(f'Cost of capital is: {coc}%')
```

```python
#USING ROE and REINVESTMENT RATE
#formula: terminal growth rate = ROE*Reinvestment rate

roe = income_statement['Net Income'] / balance_sheet['Stockholders Equity']
reinvestment_rate = 1 - ((balance_sheet['Dividends Payable'])/income_statement['Net Income'])
print(f'Reinvestment Rate is: {reinvestment_rate}')
print(f'ROE is: {roe}')
tgr = roe * reinvestment_rate
print(f'Terminal Growth Rate using ROE and Reinvestment Rate is: {tgr}')
avg_tgr = tgr.mean()
print(f'Average Terminal Growth Rate is: {avg_tgr}')
```

**THE TERMINAL GROWTH RATE USING LAST 4YRS IS 13.098%**

```python
sun=yf.Ticker("SUNPHARMA.NS")
drreddy=yf.Ticker("DRREDDY.NS")
lupin=yf.Ticker("LUPIN.NS")
aurobindo=yf.Ticker("AUROPHARMA.NS")

comps=[sun,drreddy,lupin,aurobindo,cipla]
evs=[]
ebitdas=[]
for ele in comps:
    market_cap = ele.info.get('marketCap', 0)
    cash = ele.info.get('cash', 0)
    total_debt = ele.info.get('totalDebt', 0)
    enterprise_value = market_cap+total_debt -cash
    ebitda = ele.info.get('ebitda', None)
    evs.append(enterprise_value)
    ebitdas.append(ebitda)
evs
```

```python
evs=np.array(evs)
ebitdas=np.array(ebitdas)
ev_by_ebitda=evs/ebitdas
ev_by_ebitda

multiple=ev_by_ebitda.mean()
multiple

cipla_ebitda=cipla.info.get('ebitda')
ev_cipla=cipla_ebitda*multiple
ev_cipla
```
