```python
import statsmodels.api as sm
ticker = "CIPLA.NS"
market_index = "^NSEI"

start_date = '2019-03-31'
end_date = '2024-03-31'
cipla_data= yf.download(ticker, start=start_date, end= end_date)
market_data= yf.download(market_index, start=start_date,end= end_date)
cipla_returns = cipla_data['Adj Close'].pct_change().dropna()
market_returns = market_data['Adj Close'].pct_change().dropna()

#aligning data by date
returns = pd.concat([cipla_returns, market_returns], axis=1).dropna()
returns.columns = ['Cipla', 'Market']

#REGRESSION ANALYSIS
X = returns['Market']
y = returns['Cipla']
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
beta_equity = model.params['Market']

print(f"Beta Equity for Cipla: {beta_equity:.2f}")
```

Computed Beta Equity for CIPLA = 0.40

```python
plt.figure(figsize=(10, 6))
plt.scatter(X['Market'], y, color='blue', alpha=0.5, label="Data Points")
plt.plot(X['Market'], model.predict(X), color='red', label="Regression Line")
plt.xlabel("Market Returns (NIFTY 50)")
plt.ylabel("Cipla Returns")
plt.title("Beta Calculation for Cipla using Linear Regression")
plt.legend()
plt.show()

print(model.summary())
```

```python
rf= 6.93 #risk-free rate in %
rp= 4.06 #risk-premium in %
beta_equity= 0.40

cost_of_equity_capital= rf + (beta_equity)*rp
print(f"Cost of Equity Capital for Cipla: {cost_of_equity_capital:.2f}%")
```

```python
ebit = income_statement['EBIT'].sum()
interest_expense = income_statement['Interest Expense'].sum()
icr = ebit/interest_expense #interest coverage ratio
print(f"Interest Coverage Ratio for Cipla: {icr:.2f}")
#since icr > 10, it is AAA rated and thus credit spread is 0.75%
cs = 0.75 #credit spread in %
cost_of_debt = rf + cs
print(f"Cost of Debt for Cipla: {cost_of_debt:.2f}%")
```

```python
import yfinance as yf
def get_outstanding_shares(ticker):
    stock_data = yf.Ticker(ticker)
    outstanding_shares = stock_data.info['sharesOutstanding']  # Fetch the number of shares outstanding
    return outstanding_shares

def get_stock_price(ticker):
    stock_data = yf.Ticker(ticker)
    stock_price = stock_data.history(period='1d')['Close'].iloc[-1]  # Get the latest closing price
    return stock_price

ticker = "CIPLA.NS"
outstanding_shares = get_outstanding_shares(ticker)
stock_price = get_stock_price(ticker)
mve = outstanding_shares * stock_price
print(f'Market value of equity is: {mve}')
```

```python
yield_on_debt = 0.0768  # 7.68% yield
time_to_maturity = 5  # 5 years remaining

pv_interest_payments = (income_statement['Interest Expense'] / yield_on_debt) * (1 - (1 / (1 + yield_on_debt) ** time_to_maturity))
pv_principal = balance_sheet['Total Debt'] / ((1 + yield_on_debt) ** time_to_maturity)

mvd = pv_principal + pv_interest_payments
print(f'Market Value of Debt from Book Value is: {mvd}')
avg_mvd = mvd.mean()
print(f'Average Market Value of Debt is: {avg_mvd}')
```

```python
vfirm = mve + mvd #value of firm
print(f'Value of firm is: {vfirm}')

coc = (mve/vfirm)*cost_of_equity_capital + (mvd/vfirm)*cost_of_debt*(1-tax_rate)
print(f'Cost of capital is: {coc}%')
```

**Average Cost of Capital for Cipla is 8.525%**

```python
#USING ROE and REINVESTMENT RATE
#formula: terminal growth rate = ROE*Reinvestment rate

roe = income_statement['Net Income'] / balance_sheet['Stockholders Equity']
reinvestment_rate = 1 - ((balance_sheet['Dividends Payable'])/income_statement['Net Income'])
print(f'Reinvestment Rate is: {reinvestment_rate}')
print(f'ROE is: {roe}')
tgr = roe * reinvestment_rate
print(f'Terminal Growth Rate using ROE and Reinvestment Rate is: {tgr}')
avg_tgr = tgr.mean()
print(f'Average Terminal Growth Rate is: {avg_tgr}')
```

**THE TERMINAL GROWTH RATE USING LAST 4YRS IS 13.098%**

```python
#USING THE RELATIVE VALUATION METHOD
tgrs = [0.06,0.17,0.21,0.12,0.15] #calculated by multiplying ROE and retention ratios for 5 different pharmaceutical companies
avg_tgrs = sum(tgrs)/len(tgrs)
print(f'Average Terminal Growth Rate using Relative Valuation Method is: {avg_tgrs}')
```
